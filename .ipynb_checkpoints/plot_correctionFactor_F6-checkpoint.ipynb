```python
import numpy as np
import oceanDAS as odas
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import date, datetime
from scipy import signal
import cmocean
%matplotlib inline
%cd -q '/Users/hglover/Library/CloudStorage/Box-Box/FiberExperiments/DAScomparison_paper/'

# load useful info like depth and noise floor freq for each site
h, f_noise, clrs = odas.all_line_info()
```

```python
# plot example of Duck correction factor
pname = 'DuckNC_Glover/'
with np.load(pname+'empCalNov_all.npz') as data:
    frq_duck = data['x']
    strainfac_duck = data['y'][200,:]
    chno_duck = data['ch'][200]
strainfac_duck[frq_duck>f_noise[0]]=np.nan

# plot example of KNO correction factor
pname = 'KNO_Glover/'
with np.load(pname+'empCal_ch124_p1.npz') as data:
    frq_kno = data['f']
    strainfac_kno = data['sf']
    chno_kno = 124
strainfac_kno[frq_kno>f_noise[1]]=np.nan

# plot corr factor for Marthas Vineyard
fname = 'MarthasVineyard_Smith/SpectralCorrectionFactor_channel1696_nch2112_2Hz.csv'
f = np.genfromtxt(fname, delimiter=',',skip_header=1)
frq_mv = f[:,1]
strainfac_mv = 1 / f[:,2]
strainfac_mv[frq_mv>f_noise[2]]=np.nan

# plot corr fac for Oliktok, AK
fname = 'OliktokPoint_Smith/SpectralCorrectionFactors_1Hz/SpectralCorrectionFactor_channel7960_1Hz.csv'
f = np.genfromtxt(fname, delimiter=',',skip_header=1)
frq_oli = f[:,1]
strainfac_oli = 1 / f[:,2]
strainfac_oli[frq_oli>f_noise[3]]=np.nan
# fname = 'OliktokPoint_Smith/straincorrfac7960.npz'
# with np.load(fname) as data:
#     frq_oli = data['f']
#     strainfac_oli = data['c']

# plot corr fac for Homer AK
fname = 'HomerAK_Williams/empCalex_Ch31.npz' # example for channel 31
with np.load(fname) as data:
    frq_hom = data['x']
    strainfac_hom = data['y']
strainfac_hom[frq_hom>f_noise[4]]=np.nan

# # plot example of Flo correction factor
pname = 'FlorenceOR_Ifju/'
with np.load(pname+'empCal_Ch700_Flo.npz') as data:
#     print(data.files)
    frq_flo = data['frq']
    strainfac_flo = data['sf']
    chno_flo = 700
strainfac_flo[frq_flo>f_noise[5]]=np.nan
```

```python
plt.rcParams['lines.linewidth'] = 2.5
plt.rc('font', size=12) 

plt.figure(figsize=(9,5))
plt.plot(frq_duck,strainfac_duck,'-',label='Duck, NC',c = 'gold')
plt.plot(frq_kno,strainfac_kno,'--',label='Honolulu, HI',c = clrs[1,:])
plt.plot(frq_mv,strainfac_mv,':',label='Marthas Vineyard, MA',c = clrs[2,:])
plt.plot(frq_oli,strainfac_oli,'-.',label='Oliktok Point, AK',c = clrs[3,:])
plt.plot(frq_hom,strainfac_hom,label='Homer, AK',c = clrs[4,:],linestyle=(0,(3,5,1,5)))
plt.plot(frq_flo,strainfac_flo,label='Florence, OR',c = clrs[5,:],linestyle=(0,(2,0)))

plt.legend(ncol=3,loc=(0.001,1.01))
plt.xscale('log')
plt.yscale('log')
plt.xlim([0.01,0.5])
plt.ylim([1e4,1e16])
plt.ylabel('C (-)')
plt.xlabel('Frequency (Hz)')

plt.savefig('/Users/hglover/Downloads/CorrFacs.png')
```
