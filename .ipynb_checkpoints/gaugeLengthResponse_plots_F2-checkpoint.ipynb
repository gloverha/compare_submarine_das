```python
import numpy as np
import matplotlib.pyplot as plt
from oceanDAS import dispersion
```

```python
# calc strain, pressure needs to be in units of pascals
def calcstrain(H, L, h, omega, z, t):
    x = np.arange(0, L*3, .1)
    a = H / 2
    k = 2 * np.pi / L
    x_hat = a * (np.cosh(k * (z + h)) / np.cosh(k * h)) * np.cos(k * x - omega * t)
    
    epsilon = np.gradient(x_hat) / np.gradient(x) #-a * (np.sinh(k * (z + h)) / np.sinh(k * h)) * k * np.sin(k * x - omega * t)
    
    return x, x_hat, epsilon
# calc strain, pressure needs to be in units of pascals
def estimate_gauge_strain(x, epsilon, gauge_length, x_hat):
    dx = x[1]-x[0]
    g = 9.81

    # estimate gauge strain field
    cnt = 0
    cnt2 = 0
    cnt_store = [0]
    gaugestrain = []
    gs_gl = []


    xgauge = np.arange(gauge_length/2, np.floor(x[-1]), gauge_length)
   

    for nn in range(len(x)):
        if x[nn] % gauge_length == 0 and nn != 0:
            cnt_store.append(cnt)
            cnt2 += 1
            ep = epsilon[cnt_store[cnt2-1]:cnt_store[cnt2]]
            ep_integral = np.sum(ep) * dx # integral of point strain over gauge length
            gaugestrain.append(ep_integral)
            gs_gl.append(ep_integral / gauge_length)
        cnt += 1

    if len(xgauge) > len(gaugestrain):
        xgauge = xgauge[0:-1]
        
    return gaugestrain, gs_gl, xgauge
```

```python
# calculate possible range of wave celerities

# Reasonable range of water depths
h = np.arange(5,50,1)

# IB range of wave periods
T = np.arange(4,20,1)
f = 1/T

# L,k = dispersion(h,T)
Lall = np.empty([len(h),len(T)])*np.nan
for jj in range(len(h)):
    Lall[jj,:], _ = dispersion(h[jj],T)
# L = np.arange(np.min(Lall),np.max(Lall),1)
C = Lall/T
print(np.min(C),np.max(C))

plt.figure()
p = plt.pcolor(C)
plt.colorbar(p)
```

```python
# Reasonable range of water depths
h = np.arange(5,50,0.2)

# IB range of wave periods
T = np.arange(2,20,0.2)
f = 1/T

# L,k = dispersion(h,T)
Lall = np.empty([len(h),len(T)])*np.nan
for jj in range(len(h)):
    Lall[jj,:], _ = dispersion(h[jj],T)
L = np.arange(np.min(Lall),np.max(Lall),1)
k = 1/L

# depth attenuation
atten = np.empty([len(h),len(k)])*np.nan
for jj in range(len(h)):
    atten[jj,:] = np.exp(k*h[jj])
atten = (1-1/atten)*100; # square for energy 


# Range of common gauge lengths in meters
gl = np.arange(2,20,0.25)
pw = 2e-8 # for almost all its 20 ns
hp = pw*299792458/1.46 # pulse width = 20 ns * C / n
sigma_k = 2.3548/(2*np.pi*hp)

# pre-allocate empty array
G = np.empty([len(gl),len(L)])*np.nan
P = np.empty([len(L)])*np.nan
for jj in range(len(L)):
    k = 1/L[jj]
    # calculate G using eqn 11
    G[:,jj] = np.abs(np.sin(np.pi*k*gl)) / (np.pi*k*gl)
    P[jj] = np.exp((-k**2)/(2*sigma_k**2))
H_k = G*P





# FIGURE
fig, ax = plt.subplots(2,2,figsize=(10,8))
ax = ax.ravel()
pc = ax[0].pcolor(T,h,Lall,vmin=0,vmax=300)
plt.colorbar(pc,ax=ax[0],label='Wavelength (m)')
cs = ax[0].contour(T,h,Lall,colors='k',vmin=0,vmax=300,linewidths=0.6)
ax[0].clabel(cs,levels = [100,200,300])
ax[0].set_ylabel('Water depth (m)')
ax[0].set_xlabel('Wave period (s)')
ax[0].set_title('A) Wave dispersion relation')

pc = ax[2].pcolor(L,h,atten,vmin=0,vmax=100) # ,vmin=0,vmax=150
plt.colorbar(pc,ax=ax[2],label='% Attenuated')
cs = ax[2].contour(L,h,atten,levels=np.arange(0,100,10),vmin=0,vmax=100,colors='k',linewidths=0.6)
ax[2].clabel(cs,levels=[20,40,60,80])
# ax[1].set_xlim([5,100])
ax[2].set_xlim([6,300])
ax[2].set_xscale('log')
ax[2].set_ylabel('Water depth (m)')
ax[2].set_xlabel('Wavelength (m)')
ax[2].set_title('B) Dynamic pressure attenuation')

pc = ax[3].pcolor(L,gl,H_k*100,vmin=0,vmax=100,cmap='viridis_r')
plt.colorbar(pc,ax=ax[3],label='% resolved')
cs = ax[3].contour(L,gl,H_k*100,levels=np.arange(0,100,10),vmin=0,vmax=100,colors='k',linewidths=0.6)
ax[3].clabel(cs,levels=[10,20,40,60,80])
ax[3].set_xlabel('Wavelength (m)')
ax[3].set_xlim([6,300])
ax[3].set_xscale('log')
ax[3].set_ylabel('Gauge length (m)')
ax[3].set_title('D) Gauge length interrogator sensitivity')





# use linear wave theory to create field of bottom pressures and calc integrated GL-strain
RHO = 1000
H = 1 # wave height (m)
h = 10 # depth (m)
T = 6 # wave period (s)
L, k = dispersion(h,T)
omega = 2 * np.pi / T
# plot point strain and gauge strain
gauge_length1 = np.round(L/10)
gauge_length2 = np.round(L/4)
gauge_length3 = np.round(L/2)
gauge_length4 = np.round(L/1)

x, x_hat, epsilon = calcstrain(H, L, h, omega, -h, 0)

gaugestrain1, gs_gl1, xgauge1 = estimate_gauge_strain(x, epsilon, gauge_length1, x_hat)
gaugestrain2, gs_gl2, xgauge2 = estimate_gauge_strain(x, epsilon, gauge_length2, x_hat)
gaugestrain3, gs_gl3, xgauge3 = estimate_gauge_strain(x, epsilon, gauge_length3, x_hat)
gaugestrain4, gs_gl4, xgauge4 = estimate_gauge_strain(x, epsilon, gauge_length4, x_hat)

ax[1].plot(x, epsilon,'darkgrey',linewidth = 3, label='point strain')
ax[1].scatter(xgauge1, gs_gl1,c='b')
ax[1].scatter(xgauge2, gs_gl2,c='#ff7f00')
ax[1].scatter(xgauge3, gs_gl3,c='g')
ax[1].scatter(xgauge4, gs_gl4,c='#dede00')
ax[1].plot(xgauge1, gs_gl1,c='#377eb8')
ax[1].plot(xgauge2, gs_gl2,c='#ff7f00')
ax[1].plot(xgauge3, gs_gl3,c='g')
ax[1].plot(xgauge4, gs_gl4,c='#dede00')

ax[1].scatter(xgauge1, np.ones(len(xgauge1))*0.038, marker='1',c='b') #center of gauge, '#377eb8'
ax[1].scatter(xgauge2, np.ones(len(xgauge2))*0.036, marker='1',c='#ff7f00') #center of gauge
ax[1].scatter(xgauge3, np.ones(len(xgauge3))*0.034, marker='1',c='g') #center of gauge
ax[1].scatter(xgauge4, np.ones(len(xgauge4))*0.032, marker='1',c='#dede00') #center of gauge
ax[1].set_xlabel('x (m)')
ax[1].set_xlim([0,125])
ax[1].set_ylim([-0.04,0.04])
ax[1].set_ylabel('Integrated $\epsilon$ (m/m)')
ax[1].grid()
ax[1].set_title('C) Gauge length strain sensitivity')


fig.tight_layout()
plt.savefig('/Users/hglover/Downloads/waveConsiderations.png',dpi=250)
```
